```python
from PIL import Image
import numpy as np 
import math
import matplotlib.pyplot as plt
```

```python
def getGradMag(img, kernel):
    
    gradMag = np.zeros(img.shape)
    
    s = kernel.shape[0]
    r = math.floor(s/2)
    
    #convolve
    for i in range (r,img.shape[0]-r):
        for j in range (r,img.shape[1]-r):
            val = 0
            for k in range(-r, r+1):
                for l in range(-r,r+1):
                    val += img[i+k][j+l]*kernel[k+r][l+r]

            gradMag[i][j] = val
    
    return gradMag
    
```

```python
def sharpen(img, kernel):
    
    sharpenedImg = np.zeros(shape=img.shape)
    
    grad = getGradMag(img, kernel)  

    sharpenedImg = img + (0.1*grad)
    
    return sharpenedImg
    
```

```python
img = Image.open("Cameraman.tif")
img = img.convert('L')
img = np.array(img)

kernel = np.array([ [-1,-1,-1], [-1,8,-1], [-1,-1,-1]])
s = sharpen(img, kernel)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(2,2,1)
plt.imshow(img, cmap="gray")
plt.subplot(2,2,2)
plt.imshow(s, cmap="gray")

#construct the image from the numpy array
s = s.astype('uint8')    
s = Image.fromarray(s)
s.save('Sharpened.jpg')
```
